```python
# Metric danh gia 
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score

# Ve do thi
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
y_xlmr=np.load("/content/drive/MyDrive/NLP-DS/Thực hành 1 - Sentiment analysis/y_pred_xlmr.npy")
y_pho=np.load("/content/drive/MyDrive/NLP-DS/Thực hành 1 - Sentiment analysis/y_pred_pho.npy")
y_news=np.load("/content/drive/MyDrive/NLP-DS/Thực hành 1 - Sentiment analysis/y_pred_news.npy")
y_true=np.load("/content/drive/MyDrive/NLP-DS/Thực hành 1 - Sentiment analysis/y_true.npy")
```

```python
print("==============================XLMR========================")
cf = confusion_matrix(y_true,y_xlmr)
print(cf)

evaluation = f1_score(y_true, y_xlmr, average='micro')

print("F1 - micro: " + str(evaluation))

evaluation = f1_score(y_true, y_xlmr, average='macro')
print("F1 - macro: " + str(evaluation))
```

```python
print("==============================PhoBERT========================")
cf = confusion_matrix(y_true,y_pho)
print(cf)

evaluation = f1_score(y_true, y_pho, average='micro')

print("F1 - micro: " + str(evaluation))

evaluation = f1_score(y_true, y_pho, average='macro')
print("F1 - macro: " + str(evaluation))
```

```python
print("==============================BERT4NEWs========================")
cf = confusion_matrix(y_true,y_news)
print(cf)

evaluation = f1_score(y_true, y_news, average='micro')

print("F1 - micro: " + str(evaluation))

evaluation = f1_score(y_true, y_news, average='macro')
print("F1 - macro: " + str(evaluation))
```
